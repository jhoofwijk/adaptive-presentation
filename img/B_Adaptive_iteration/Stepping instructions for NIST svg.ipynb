```python

def k(T):
    a = 1.8743
    b = -0.41538
    c = -0.6018
    d = 0.13294
    e = 0.26426
    f = -0.0219
    g = -0.051276
    h = 0.0014871
    i = 0.003723

    return (10 ** (( a + c * T ** 0.5 + e * T + g * T ** 1.5 + i*T** 2 ) / ( 1+ b*T** 0.5 + d * T + f * T ** 1.5 + h * T ** 2)))
```

```python
def clearTicks():
    return
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=False) # labels along the bottom edge are off
    plt.tick_params(
        axis='y',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        left=False,      # ticks along the bottom edge are off
        right=False,         # ticks along the top edge are off
        labelleft=False) # labels along the bottom edge are off
```

```python
l = adaptive.Learner1D(k, [4, 300])

datapoints = []
new_points = []


(x, ), _ = l.ask(1)
l.tell(x, k(x))


fig = plt.figure(figsize=(5,5))

for i in range(11):
    l._recompute_losses_factor = 1.01
    plt.cla()
    ax = plt.gca()
    (x, ), _ = l.ask(1)
    l.tell(x, k(x))
    print(l.npoints)

    new_points.append((x, k(x)))
    xx, yy = (zip(*sorted(l.data.items())))


    ax.cla()
    ax.scatter(x, k(x), color='r', zorder=100)
    ax.plot(xx,yy, marker='o')
    plt.xlabel("x")
    plt.ylabel("y")
    ymin, ymax = min(yy), max(yy)
    plt.ylim(ymin - 0.05*(ymax-ymin), ymax+ 0.05*(ymax-ymin))
    clearTicks()
    plt.savefig(f'img_{l.npoints}.svg', bbox_inches='')

    
   

    ax.cla()
    ax.plot(xx,yy, marker='o')
    plt.xlabel("x")
    plt.ylabel("y")
    xs = l.losses.items()[0][0]
    ys = k(np.array(xs))
    ax.plot(xs,ys, color='red')
    ymin, ymax = min(yy), max(yy)
    plt.ylim(ymin - 0.05*(ymax-ymin), ymax+ 0.05*(ymax-ymin))
    clearTicks()
    plt.savefig(f'img_{l.npoints}_marked.svg', bbox_inches='')
    
    
    
    ax.cla()
    ax.plot(xx,yy, marker='o')
    plt.xlabel("x")
    plt.ylabel("y")
    xs = l.losses.items()[0][0]
    ys = k(np.array(xs))
    ax.plot(xs,ys, color='red')
    
    ax.annotate('', xy=(np.average(xs), 1.05*min(yy)-0.05*max(yy)), xytext=(np.average(xs), 1.15*min(yy)-0.15*max(yy)),
                arrowprops=dict(facecolor='black', shrink=0.05),
                )
    
    ymin, ymax = min(yy), max(yy)
    plt.ylim(ymin - 0.05*(ymax-ymin), ymax+ 0.05*(ymax-ymin))
    clearTicks()
    plt.savefig(f'img_{l.npoints}_marked_2.svg', bbox_inches='')
    
    

```
