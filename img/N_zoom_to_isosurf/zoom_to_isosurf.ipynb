```python
import adaptive
adaptive.notebook_extension()

from adaptive.learner.learnerND import curvature_loss_function
from adaptive.learner.base_learner import uses_nth_neighbors
import numpy as np
```

```python
N = 25
```

```python

normal_loss = curvature_loss_function(0.05)

level = 0

@uses_nth_neighbors(1)
def loss(simplex, ys, nei, neiy):
    l = normal_loss(simplex, ys, nei, neiy)
    if min(ys) < level and max(ys) > level:
        return l * 4
    return l

def f(xy):
    x, y = xy
    return x**2 + y**3 - 0.5

def f(xy):
    x, y = xy
    return x**2 + y**3 - 0.5 + 2 * np.exp(-10 * (x**2 + (y+.7)**2))


learner = adaptive.LearnerND(f, bounds=[(-1,1)]*2, loss_per_simplex=loss)

runner = adaptive.BlockingRunner(learner, goal=lambda l: l.npoints >= N**2)
```

```python
learner.plot_isoline(0, n=200, tri_alpha=0.7, opts=dict(color='black', line_width=2)).options(height=500, width=500)
```
