```python
import adaptive
adaptive.notebook_extension()

from adaptive.learner.learnerND import curvature_loss_function, default_loss
from adaptive.learner.base_learner import uses_nth_neighbors
import numpy as np
```

```python
N = 25
```

```python

normal_loss = curvature_loss_function(0.05)

level = 0

@uses_nth_neighbors(1)
def loss(simplex, ys, nei, neiy):
    l = normal_loss(simplex, ys, nei, neiy)
    if min(ys) < level and max(ys) > level:
        return l * 5
    return l

def f(xy):
    x, y = xy
    return x**2 + y**2 - 0.5

learner = adaptive.LearnerND(f, bounds=[(-1,1)]*2, loss_per_simplex=loss)

runner = adaptive.BlockingRunner(learner, goal=lambda l: l.npoints >= N**2)
```

```python
learner.plot_isoline(0, n=200, tri_alpha=0.7, opts=dict(color='red', line_width=2)).options(height=500, width=500)
```

```python

normal_loss = curvature_loss_function(0.05)

level = 0

def trans(ys):
    r = []
    for y in ys:
        if y is None:
            r.append(None)
        else:
            r.append(np.tanh(25*y))
    return r

def loss(simplex, ys):
    l = default_loss(simplex, trans(ys))
    return l

def f(xy):
    x, y = xy
    return x**2 + y**2 - 0.5

learner = adaptive.LearnerND(f, bounds=[(-1,1)]*2, loss_per_simplex=loss)

runner = adaptive.BlockingRunner(learner, goal=lambda l: l.npoints >= N**2)
learner.plot(tri_alpha=0.7)
```

```python
learner.plot_isoline(0, n=200, tri_alpha=0.7, opts=dict(color='red', line_width=2)).options(height=500, width=500)
```
